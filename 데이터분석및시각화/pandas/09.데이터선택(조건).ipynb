# 09. 데이터선택(조건)
   조건에 해당하는 데이터 선택

```python
import pandas as pd
df = pd.read_excel('score.xlsx',index_col='지원번호')
df
```

```python
df['키'] >= 185 # 학생들의 키가 185 이상인지 여부를 True / False
```

```python
filt = (df['키'] >= 185)
df[filt]
```

```python
df.loc[df['키'] >= 185, '수학'] # 키가 185 이상인 학생들의 수학 데이터
```

```python
df.loc[df['키'] >= 185, ['이름', '수학', '과학']]
# 키가 185 이상인 학생들의 이름, 수학, 과학 데이터
```

## 다양한 조건
   & 그리고

```python
# 키가 185 이상인 북산고 학생 데이터
df.loc[(df['키'] >= 185) & (df['학교'] == '북산고') ]
```

### | 또는

```python
# 키가 170보다 작거나 키가 200보다 큰 학생 데이터
df.loc[(df['키'] < 170) | (df['키'] > 200)]
```

### str 함수

```python
filt = df['이름'].str.startswith('송') # '송'씨 성을 가진 사람
df[filt]
```

```python
filt = df['이름'].str.contains('태') # 이름에 '태'가 들어간 사람
df[filt]
```

```python
df[~filt] # 이름에 '태'가 들어가지 않는 사람
```

```python
langs = ["Python", "Java"]
filt = df['SW특기'].isin(langs) # SW특기가 Python이거나 Java인 사람
df[filt]
```

```python
langs = ["python", "java"]
filt = df['SW특기'].str.lower().isin(langs)
df[filt]
```

```python
filt = df['SW특기'].str.contains('Java')
df[filt]
```

```python
# NaN 데이터에 대해서 True로 설정
df['SW특기'].str.contains('Java', na=True)
```

```python
# NaN 데이터에 대해서 True로 설정
filt = df['SW특기'].str.contains('Java', na=False)
df[filt]
```
