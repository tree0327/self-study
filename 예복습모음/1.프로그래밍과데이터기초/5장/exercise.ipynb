````python
from datetime import timedelta

def calculate_working_days(start_date, end_date, holidays=None):
    # ê³µíœ´ì¼ì´ ì—†ìœ¼ë©´ ë¹ˆ ë¦¬ìŠ¤íŠ¸
    if holidays is None:
        holidays = []
        
    # 1. ë¬¸ìì—´ -> ë‚ ì§œ ê°ì²´ ë³€í™˜
    start = datetime.strptime(start_date, "%Y-%m-%d")
    end = datetime.strptime(end_date, "%Y-%m-%d")
    
    # 2. ì´ ì¼ìˆ˜ ê³„ì‚° (ëë‚˜ëŠ” ë‚  - ì‹œì‘í•˜ëŠ” ë‚  + 1)
    # ì‹œì‘í•˜ëŠ” ë‚ ë„ í¬í•¨ (+1)
    total_days = (end - start).days + 1
    
    weekend_count = 0  # ì£¼ë§
    holiday_count = 0  # ê³µíœ´ì¼
    working_days = 0   # ì‹¤ì œ ì¼í•˜ëŠ” ë‚ 
    
    # 3. í•˜ë£¨ì”© í™•ì¸
    for i in range(total_days):
        # í˜„ì¬ í™•ì¸í•˜ê³  ìˆëŠ” ë‚ ì§œ = ì‹œì‘ì¼ + iì¼
        current_day = start + timedelta(days=i)
        
        # ë‚ ì§œë¥¼ ë¬¸ìì—´ë¡œ ë³€ê²½
        current_day_str = current_day.strftime("%Y-%m-%d")
        
        # weekday(): 0 ~ 6ìœ¼ë¡œ ìš”ì¼ í‘œì‹œ
        is_weekend = current_day.weekday() >= 5 # ì£¼ë§ì´ë©´ True
        is_holiday = current_day_str in holidays
        
        if is_weekend:
            weekend_count += 1
        elif is_holiday:
            # ê³µíœ´ì¼ì´ë©´ í‰ì¼ and ê³µíœ´ì¼
            holiday_count += 1
        else:
            # ì¼í•˜ëŠ” ë‚ 
            working_days += 1
            
    return {
        "ì´ ì¼ìˆ˜": total_days,
        "ì£¼ë§ ìˆ˜": weekend_count,
        "ê³µíœ´ì¼ ìˆ˜": holiday_count,
        "ì‹¤ì œ ê·¼ë¬´ì¼": working_days
    }

# 2024ë…„ 5ì›” 1ì¼ ~ 5ì›” 10ì¼ (ì–´ë¦°ì´ë‚  í¬í•¨)
holidays_2024 = get_holidays_of_2024() # ìœ„ì—ì„œ ë§Œë“  ê³µíœ´ì¼ ì‚¬ì „
result = calculate_working_days("2024-05-01", "2024-05-10", holidays_2024)

print(f"ì´ ê¸°ê°„: {result['ì´ ì¼ìˆ˜']}ì¼")
print(f"ì£¼ë§: {result['ì£¼ë§ ìˆ˜']}ì¼")
print(f"ê³µíœ´ì¼(í‰ì¼): {result['ê³µíœ´ì¼ ìˆ˜']}ì¼")
print(f"ì‹¤ì œ ê·¼ë¬´ì¼: {result['ì‹¤ì œ ê·¼ë¬´ì¼']}ì¼")
````

í”„ë¡œì íŠ¸ ì‹œì‘ì¼ê³¼ ì†Œìš” ê¸°ê°„(ê·¼ë¬´ì¼ ê¸°ì¤€)ì„ ì…ë ¥ë°›ì•„ ì™„ë£Œ ì˜ˆì •ì¼ì„ ê³„ì‚°í•˜ëŠ” í•¨ìˆ˜ë¥¼ ì‘ì„±í•˜ì„¸ìš”.

```python
def calculate_project_schedule(start_date, duration_days, holidays=None):
    """
    í”„ë¡œì íŠ¸ ì‹œì‘ì¼ê³¼ ì†Œìš” ê¸°ê°„ìœ¼ë¡œ ì™„ë£Œì¼ì„ ê³„ì‚°í•©ë‹ˆë‹¤.
    
    Args:
        start_date (str): í”„ë¡œì íŠ¸ ì‹œì‘ì¼
        duration_days (int): ì†Œìš” ê·¼ë¬´ì¼
        holidays (list): ê³µíœ´ì¼ ë¦¬ìŠ¤íŠ¸

        dict: í”„ë¡œì íŠ¸ ì¼ì • ì •ë³´

```

**ì¶œë ¥ ì˜ˆì‹œ:**
- í”„ë¡œì íŠ¸ ì™„ë£Œ ì˜ˆì •ì¼
- ì†Œìš” ê·¼ë¬´ì¼
- ì´ ë‹¬ë ¥ ì¼ìˆ˜
- í”„ë¡œì íŠ¸ ê¸°ê°„ ì¤‘ ê³µíœ´ì¼ ìˆ˜
- ë§ˆì¼ìŠ¤í†¤ ë‚ ì§œë“¤ (25%, 50%, 75% ì§„í–‰ ì‹œì )

````python
def calculate_project_schedule(start_date, duration_days, holidays=None):
    if holidays is None:
        holidays = []
        
    start = datetime.strptime(start_date, "%Y-%m-%d")
    
    current_day = start
    days_worked = 0       # ì§€ê¸ˆê¹Œì§€ ì¼í•œ ë‚ ì§œ ìˆ˜
    holiday_count = 0     # ì‰¬ì—ˆë˜ ê³µíœ´ì¼ ìˆ˜
    calendar_days = 0     # ë‹¬ë ¥ìƒ ì§€ë‚˜ê°„ ì´ ì¼ìˆ˜
    
    # ëª©í‘œ ê·¼ë¬´ì¼ë§Œí¼ ë°˜ë³µ
    while days_worked < duration_days:
        # 1. ì˜¤ëŠ˜ì´ ì£¼ë§ì¸ì§€ í™•ì¸
        is_weekend = current_day.weekday() >= 5
        
        # 2. ì˜¤ëŠ˜ì´ ê³µíœ´ì¼ì¸ì§€ í™•ì¸
        current_day_str = current_day.strftime("%Y-%m-%d")
        is_holiday = current_day_str in holidays
        
        # not ì£¼ë§, not ê³µíœ´ì¼ -> ì¼í•œê²ƒ
        if not is_weekend and not is_holiday:
            days_worked += 1
        elif is_holiday and not is_weekend:
            # í‰ì¼ì¸ë° ì‰° ë‚ 
            holiday_count += 1
            
        # ëª©í‘œë¥¼ ë‹¬ì„±í•˜ì§€ ëª»í–ˆìœ¼ë©´ ë‹¤ìŒ ë‚ ë¡œ
        if days_worked < duration_days:
            current_day += timedelta(days=1)
            calendar_days += 1
            
    # ë§ˆì¼ìŠ¤í†¤ ë‚ ì§œ ê³„ì‚°í•˜ê¸° (ì¤‘ê°„ì ê²€)
    # ì „ì²´ ê¸°ê°„ì˜ 25%, 50%, 75% ì§€ì  ê³„ì‚°
    milestones = {}
    for percent in [25, 50, 75]:
        # ì „ì²´ ë‹¬ë ¥ ì¼ìˆ˜ * í¼ì„¼íŠ¸
        days_passed = int(calendar_days * (percent / 100))
        milestone_date = start + timedelta(days=days_passed)
        milestones[f"{percent}%"] = milestone_date.strftime("%Y-%m-%d")
    
    return {
        "ì™„ë£Œ ì˜ˆì •ì¼": current_day.strftime("%Y-%m-%d"),
        "ì†Œìš” ê·¼ë¬´ì¼": duration_days,
        "ì´ ë‹¬ë ¥ ì¼ìˆ˜": calendar_days + 1, # ë§ˆì§€ë§‰ ë‚  í¬í•¨
        "ê³µíœ´ì¼ ìˆ˜": holiday_count,
        "ë§ˆì¼ìŠ¤í†¤": milestones
    }

# 10ì¼ë™ì•ˆ ì¼í•´ì•¼ í•˜ëŠ” í”„ë¡œì íŠ¸ë¥¼ 2024ë…„ 5ì›” 1ì¼ì— ì‹œì‘í•˜ë©´ ì–¸ì œ ëë‚ ê¹Œìš”?
proj = calculate_project_schedule("2024-05-01", 10, holidays_2024)

print(f"ì‹œì‘ì¼: 2024-05-01")
print(f"ì™„ë£Œì¼: {proj['ì™„ë£Œ ì˜ˆì •ì¼']}")
print(f"ì´ ê±¸ë¦° ê¸°ê°„: {proj['ì´ ë‹¬ë ¥ ì¼ìˆ˜']}ì¼ (ì£¼ë§/íœ´ì¼ í¬í•¨)")
print(f"ì¤‘ê°„ ì ê²€ì¼: {proj['ë§ˆì¼ìŠ¤í†¤']}")
````
